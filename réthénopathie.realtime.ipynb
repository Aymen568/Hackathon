```python

import cv2
import numpy as np
from keras.models import load_model

model = load_model('C:/Users/DELL/Desktop/keras_model (3).h5')

cap= cv2.VideoCapture(0)

labels = open('C:/Users/DELL/Desktop/labels (3).txt', 'r').readlines()

while True:

    ret, image = cap.read()

    im = cv2.resize(image, (224, 224), interpolation=cv2.INTER_AREA)
   
    im = np.asarray(im, dtype=np.float32).reshape(1, 224, 224, 3)

    im = (im / 127.5) - 1

   
    prediction = model.predict(im)
    index = np.argmax(prediction)
    class_name = labels[index]
    confidence_score = prediction[0][index]
    keyboard_input = cv2.waitKey(1)
   
    print('Class:', class_name, end='')
    print('Confidence score:', confidence_score)
   
    cv2.putText(image,class_name,(20,50), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 255, 255), 2)
   
    cv2.imshow('Webcam Image', image)
   
    if keyboard_input == 27:
        break

camera.release()
cv2.destroyAllWindows()
```
